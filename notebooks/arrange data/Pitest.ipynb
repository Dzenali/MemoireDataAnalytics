```python
import pandas as pd
import os

reports_path = "../../generated/reports-no-bug"
code_analysis_generated_path = "../../generated/code-analysis"

if not os.path.exists(code_analysis_generated_path):
    os.makedirs(code_analysis_generated_path)

data = []

users = os.listdir(reports_path)
for user in users:
    uploads_path = os.path.join(reports_path, user)

    pitest_path = os.path.join(uploads_path, "Pitest", "mutations.xml")
    pitest_valid = os.path.exists(pitest_path)

    data.append([user, pitest_path, pitest_valid])


main_df = pd.DataFrame(data, columns=['user', 'pitest', 'pitest_valid'])

main_df
```

# Add group

```python
users_csv_path = '../../generated/database/users.csv'
if not os.path.exists(users_csv_path):
    raise FileNotFoundError(f"Le fichier {users_csv_path} n'existe pas. Exécutez le notebook 'notebooks/arrange data/Database data.ipynb' pour le générer.")

df_users = pd.read_csv(users_csv_path, usecols=['user', 'game_mode'])
main_df = main_df.merge(df_users, on=['user'], how='left').dropna()
main_df 
```

# Invalid data

```python
invalid_pitest = main_df[main_df["pitest_valid"] == False]

print(f"Nombre total de projets : {len(main_df)}")
print(f"Nombre de projets valides Pitest :", f"{len(main_df) - len(invalid_pitest)}/{len(main_df)}", f"({(len(main_df) - len(invalid_pitest)) / len(main_df) * 100:.2f}%)")
print(f"Nombre de projets invalides Pitest :", f"{len(invalid_pitest)}/{len(main_df)}", f"({len(invalid_pitest) / len(main_df) * 100:.2f}%)")
print(f"\t Groupe SOLO : {len(invalid_pitest[invalid_pitest['game_mode'] == 'SOLO'])}")
print(f"\t Groupe TEAM : {len(invalid_pitest[invalid_pitest['game_mode'] == 'TEAM'])}")
```

# User both sessions Pitest valid

```python
cols = ['instruction', 'branch', 'line', 'method']
users_count = len(main_df.groupby('user'))
users_with_valid_pitest = main_df[(main_df['pitest_valid'] == True)]
users = main_df[main_df['user'].isin(users_with_valid_pitest['user'])].copy()

print(f"Nombre d'utilisateurs Pitest valide : {len(users)}/{users_count}")
users
```

```python
users_pitest_data = []

for _, user_data in users.iterrows():
    user_pitest_df = pd.read_xml(user_data['pitest']).drop(columns=['indexes', 'blocks', 'killingTests', 'succeedingTests', 'coveringTests'])

    users_pitest_data.append(pd.concat([
        pd.DataFrame({
            'user': user_data['user'],
            'game_mode': user_data['game_mode'],
        }, index=user_pitest_df.index),
        user_pitest_df
    ], axis=1))

pitest_data = pd.concat(users_pitest_data, ignore_index=True)
pitest_data
```

```python
df = pitest_data.groupby(['user', 'game_mode', 'sourceFile', 'detected']).size().unstack(fill_value=0)
df.rename(columns={False: 'survived', True: 'killed'}, inplace=True)
df['mutants_count'] = df['survived'] + df['killed']
df.to_csv(f"{code_analysis_generated_path}/pitest-details.csv", index=True)
df
```

```python
project_data_df = df.groupby(['user', 'game_mode']).sum()
project_data_df['mutation_coverage'] = project_data_df['killed'] / project_data_df['mutants_count']
project_data_df = project_data_df.reset_index()
project_data_df.to_csv(f'{code_analysis_generated_path}/pitest.csv', index=False)
project_data_df
```
