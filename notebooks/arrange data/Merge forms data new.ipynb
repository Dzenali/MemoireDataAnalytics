```python
import os
import pandas as pd
csv_files_path = "../../data/forms"

forms_generated_path = "../../generated/forms"
if not os.path.exists(forms_generated_path):
    os.makedirs(forms_generated_path)

output_csv_file = f"{forms_generated_path}/data.csv"
invalid_data_csv_file = f"{forms_generated_path}/invalid-data.csv"

df_team = pd.read_csv(f"{csv_files_path}/TEAM.csv")
df_solo = pd.read_csv(f"{csv_files_path}/SOLO.csv")
df_demo = pd.read_csv(f"{csv_files_path}/demographic.csv")

# columns to take from each csv file
demo_cols = [
    'ID',
    'En quelle année d\'étude êtes vous ?',
    'Quels outils gamifiés avez vous déjà utilisé ?',
    'Comment évaluez-vous votre connaissance de Java?',
    'Comment évaluez-vous votre compétence en tests logiciels ?',
    'Comment évaluez-vous votre motivation à écrire des tests logiciels ?'
]

sus_cols = [
    'ID',
    'Je voudrais utiliser ce système fréquemment.',
    'Ce système est inutilement complexe.',
    'Ce système est facile à utiliser.',
    'J\'aurais besoin du soutien d\'un technicien pour être capable d\'utiliser ce système.',
    'Les différentes fonctionnalités de ce système sont bien intégrées.',
    'Il y a trop d\'incohérences dans ce système.',
    'La plupart des gens apprendront à utiliser ce système très rapidement.',
    'Ce système est très lourd à utiliser.',
    'Je me suis senti·e très confiant·e en utilisant ce système.',
    'J\'ai eu besoin d\'apprendre beaucoup de choses avant de pouvoir utiliser ce système.'
]

commun_cols = [
    'ID',
    'Ennui',
    'Stimulation',
    'Frustration',
    'Accomplissement',
    'Fatigue',
    'J\'ai été distrait par le mode de jeu pendant que j\'écrivais mes tests.',
    'J\'étais plus motivé à écrire des tests avec le mode de jeu que d\'habitude.',
    'Le système de mode de jeu m\'incite à écrire plus de tests.',
    'Mon engouement pour un système gamifié avec uniquement un mode de jeu diminuerait au file du temps (plusieurs semaines/mois/années) (sentiment de lassitude).',
    'Les succès donnaient des objectifs clairs.',
    'Le feedback des succès était suffisant pour connaitre mon avancement.',
    'mode'
]

special_team_cols = [
    'ID',
    'Avoir des succès d\'équipe vous a.',
    'Dépendre des autres pour débloquer les succès vous a.',
    'Savoir que vous deviez débloquer un palier pour permettre à vos coéquipiers d\'accéder à des succès vous a.',
    'Il y avait une ambiance de compétition dans l\'équipe.'
]

# select only the columns we need and merge the dataframes
demo = df_demo[demo_cols]
sus_team    = df_team[sus_cols]
sus_solo    = df_solo[sus_cols]
commun_team = df_team[commun_cols]
commun_solo = df_solo[commun_cols]
special_team = df_team[special_team_cols]

sus = pd.concat([sus_team, sus_solo], ignore_index=True)
commun = pd.concat([commun_team, commun_solo], ignore_index=True)

data = pd.merge(demo, sus, on='ID', how='outer')
data = pd.merge(data, commun, on='ID', how='outer')
data = pd.merge(data, special_team, on='ID', how='outer')

filtered_data = data[data['mode'].notna()]
invalid_data = data[data['mode'].isna()]
```

```python
filtered_data.to_csv(output_csv_file, index=False)
print(f"Filtered data merged and saved to {output_csv_file}")

invalid_data.to_csv(invalid_data_csv_file, index=False)
print(f"Invalid data merged and saved to {invalid_data_csv_file}")
```
