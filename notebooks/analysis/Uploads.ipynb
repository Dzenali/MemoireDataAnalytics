```python
import pandas as pd
import os

reports_path = "../../generated/reports-no-bug"

data = []

users = os.listdir(reports_path)
for user in users:
    uploads_path = os.path.join(reports_path, user)

    jacoco_path = os.path.join(uploads_path, "JaCoCo", "jacoco.csv")
    jacoco_valid = os.path.exists(jacoco_path)

    pitest_path = os.path.join(uploads_path, "Pitest", "mutations.xml")
    pitest_valid = os.path.exists(pitest_path)

    data.append([user, uploads_path, jacoco_path, jacoco_valid, pitest_path, pitest_valid])

main_df = pd.DataFrame(data, columns=['user', 'uploads','jacoco', 'jacoco_valid', 'pitest', 'pitest_valid'])
```

# Add group

```python
users_csv_path = '../../generated/database/users.csv'
if not os.path.exists(users_csv_path):
    raise FileNotFoundError(f"Le fichier {users_csv_path} n'existe pas. Exécutez le notebook 'notebooks/arrange data/Database data.ipynb' pour le générer.")

df_users = pd.read_csv(users_csv_path, usecols=['user', 'game_mode']).dropna()
main_df = main_df.merge(df_users, on='user', how='left')
main_df
```

# Invalid data

```python
invalid_jacoco = main_df[main_df["jacoco_valid"] == False]
invalid_pitest = main_df[main_df["pitest_valid"] == False]
both_invalid = main_df[(main_df["jacoco_valid"] == False) | (main_df["pitest_valid"] == False)]

both_invalid.to_csv("../../generated/database/invalid_uploads.csv", index=False)

print(f"Nombre total d'utilisateurs : {len(df_users)}")
print(f"Nombre total de projets : {len(main_df)}")
print("")

print(f"Nombre de projets invalides JaCoCo :", len(invalid_jacoco), f"({len(invalid_jacoco) / len(main_df) * 100:.2f}%)")
print(f"\t Groupe solo : {len(invalid_jacoco[invalid_jacoco['game_mode'] == 'solo'])}")
print(f"\t Groupe équipe : {len(invalid_jacoco[invalid_jacoco['game_mode'] == 'équipe'])}")
print("")

print(f"Nombre de projets invalides Pitest :", len(invalid_pitest), f"({len(invalid_pitest) / len(main_df) * 100:.2f}%)")
print(f"\t Groupe solo : {len(invalid_pitest[invalid_pitest['game_mode'] == 'solo'])}")
print(f"\t Groupe équipe : {len(invalid_pitest[invalid_pitest['game_mode'] == 'équipe'])}")
print("")

print(f"Nombre de projets invalide (au moins 1 des 2) :", len(both_invalid), f"({len(both_invalid) / len(main_df) * 100:.2f}%)")
print(f"\t Groupe solo : {len(both_invalid[both_invalid['game_mode'] == 'solo'])}")
print(f"\t Groupe équipe : {len(both_invalid[both_invalid['game_mode'] == 'équipe'])}")
```
